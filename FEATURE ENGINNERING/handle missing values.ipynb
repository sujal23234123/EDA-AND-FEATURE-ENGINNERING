#FILLING THOSE MISSING VALUES USING MEAN 

```python
dataset2["agemean"]=dataset2["age"].fillna(dataset2["age"].mean())
dataset2[["agemean","age"]]
```

```python
dataset2.head()
```

```python
sns.displot(dataset2["agemean"],kde=True)
```

#USING MEDIAN TO FILL THOSE MISSING VALUES


```python
dataset3=sns.load_dataset("titanic")
dataset3.head()
```

```python
dataset3.isnull()
```

```python
dataset3.isnull().sum()
```

```python
dataset3["agemedian"]=dataset3["age"].fillna(dataset3["age"].median())
dataset3[["agemedian","age"]]


```

```python
sns.displot(dataset3["agemedian"],kde=True)
```

```python
dataset3.head()
```

```python
dataset3[dataset3["embarked"].isnull()]
```

```python
embarked_mode=dataset3[dataset3["embarked"].notna()]["embarked"].mode()[0]
dataset3["embarked_mode"]=dataset3["embarked"].fillna(embarked_mode)
```
